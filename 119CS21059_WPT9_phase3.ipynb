## 1.    Univariate Analysis 

Univariate analysis for continuous variable
There are two methods
1 Tabular method : mean, median, std, missing values using describe()

2 Graphical method : distribution, outliers

using box plot for outlies 

histogram for distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a DataFrame named 'data' with a 'price' column.
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['price'], color='skyblue')
plt.title('Box Plot of Airbnb Listing Prices in New York')
plt.xlabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=30, kde=True, color='skyblue')
plt.title('Histogram of Airbnb Listing Prices in New York')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

## 2.    Bivariate Analysis 

Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=data, x='room_type', y='price')
plt.title('Scatter Plot of Price vs. Number of Bedrooms')
plt.show()
```

## 3.    Multivariate Analysis 

```python
import matplotlib.pyplot as plt
# Let's consider price, number_of_reviews, calculated_host_listings_count, and availability_365 for analysis
selected_columns = ['price', 'number_of_reviews', 'calculated_host_listings_count', 'availability_365']

# Create a subset with selected columns
subset = data[selected_columns]

# Pairwise scatterplot to visualize relationships
pd.plotting.scatter_matrix(subset, figsize=(12, 12))
plt.suptitle('Pairwise Scatterplot of Selected Variables')
plt.show()
```

```python
# Correlation matrix to quantify relationships
correlation_matrix = subset.corr()
print('Correlation Matrix:')
print(correlation_matrix)

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Correlation Heatmap')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation='vertical')
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.show()
```

## 4.    Visualize above with Bar, Box, FacetGrid, Heatmap, Crosstab, Line plot Submission ( Python File with dataset) 

### 1. Bar Plot: 

```python
import matplotlib.pyplot as plt

# Bar plot for neighborhood_group
data['neighbourhood_group'].value_counts().plot(kind='bar')
plt.title('Distribution of Airbnb Listings by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Count')
plt.show()
```

### 2. Box Plot: 

```python
import seaborn as sns

# Box plot for price by room type
sns.boxplot(x='room_type', y='price', data=data)
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

### 3. FacetGrid: 

```python
# FacetGrid for price by neighborhood_group
g = sns.FacetGrid(data, col='neighbourhood_group')
g.map(sns.histplot, 'price')
g.set_xlabels('Price')
g.set_titles('Neighborhood Group: {col_name}')
plt.show()
```

### 4. Heatmap: 

```python
# Correlation heatmap for selected variables
corr_matrix = data[['price', 'number_of_reviews', 'calculated_host_listings_count', 'availability_365']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### 5. Crosstab:

```python
# Crosstab between neighborhood_group and room_type
crosstab = pd.crosstab(data['neighbourhood_group'], data['room_type'])
crosstab.plot(kind='bar', stacked=True)
plt.title('Crosstab: Neighborhood Group vs. Room Type')
plt.xlabel('Neighborhood Group')
plt.ylabel('Count')
plt.show()
```

### 6. Line Plot:

```python
# Line plot of reviews_per_month over time
data['last_review'] = pd.to_datetime(data['last_review'])
data['month'] = data['last_review'].dt.month
reviews_by_month = data.groupby('month')['reviews_per_month'].mean()
reviews_by_month.plot(kind='line', marker='o')
plt.title('Average Reviews per Month Over Time')
plt.xlabel('Month')
plt.ylabel('Average Reviews per Month')
plt.show()
```
